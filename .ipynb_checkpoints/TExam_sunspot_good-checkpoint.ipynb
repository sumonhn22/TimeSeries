```python
import csv
import numpy as np
time_step = []
sunspots = []

with open('Sunspots.csv') as csvfile:
  reader = csv.reader(csvfile, delimiter=',')
  next(reader)
  for row in reader:
    sunspots.append(float(row[2]))
    time_step.append(int(row[0]))

# print(sunspots)
series = np.array(sunspots)
time = np.array(time_step)
```

```python
def normalize_series(data, min, max):
    data = data - min
    data = data / max
    return data

def plot_series(time, series, format="-", start=0, end=None):
    plt.plot(time[start:end], series[start:end], format)
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.grid(True)
```

```python
series = normalize_series(series, series.min(axis=0), series.max(axis=0))
# print(series)

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()

n_features = 1
```

```python
split_time = 3000
# print(len(series))
time_train, x_train = time[:split_time], series[:split_time]
time_valid, x_valid = time[split_time:], series[split_time:]
split_time
```

```python
########################     window   ########################
def windowed_dataset(series, batch_size, n_past, n_future, shift):
    ds = tf.data.Dataset.from_tensor_slices(series)
    ds = ds.window(size=n_past + n_future, shift=shift, drop_remainder=True)
    ds = ds.flat_map(lambda w: w.batch(n_past + n_future))
    ds = ds.map(lambda w: (w[:n_past], w[n_past:]))
    return ds.batch(batch_size).prefetch(1)
```

```python
n_past=10
n_future=10
batch_size=20
train_set = windowed_dataset(series=x_train, batch_size=batch_size, n_past=n_past, n_future=n_future, shift=1)
valid_set = windowed_dataset(series=x_valid, batch_size=batch_size, n_past=n_past, n_future=n_future, shift=1)
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Conv1D(
             filters=64, kernel_size=5, strides=1, 
             padding='causal', activation='relu', 
             input_shape=[n_past, 1]), 
  tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=True)),
  tf.keras.layers.Dense(16, activation="relu"),
  tf.keras.layers.Dense(1) 
])
model.summary()
```

```python
model.compile(loss="mse", optimizer='adam', metrics=["mae"])
history = model.fit(train_set, epochs=5, verbose= 1) 
```

```python
# Plotting mae values
plt.plot(history.history['mae'])
plt.title('Model Mae')
plt.ylabel('Mae')
plt.xlabel('Epoch')
plt.show()

plt.plot(history.history['loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```

```python
def model_forecast(model, ds, window_size, batch_size):
  # ds = tf.expand_dims(ds, axis=-1)
   ds = tf.data.Dataset.from_tensor_slices(ds)
   ds = ds.window(window_size, shift=1, drop_remainder=True)
   ds = ds.flat_map(lambda w: w.batch(window_size))
   ds = ds.batch(batch_size, drop_remainder=True).prefetch(1)
 #  print('data before predict=', ds)
   ds = model.predict(ds)
   return ds

rnn_forecast = model_forecast(model, series, n_past, batch_size)
print('original forecast shape=', rnn_forecast.shape)     # forecast shape= (3200, 10, 1)
   
rnn_forecast = rnn_forecast[3000-10:-1, 0, 0]      # 3200-(3000-10)=210 
print('modified forecast =', rnn_forecast.shape)


x_valid2 = np.squeeze(x_valid[:rnn_forecast.shape[0]])
time_valid = np.squeeze(time_valid[:rnn_forecast.shape[0]])

print('modified x_valid=', x_valid2.shape)

result = tf.keras.metrics.mean_absolute_error(x_valid2, rnn_forecast).numpy()
print(result)
```

```python
plt.figure(figsize=(20, 6))
plot_series(time_valid, x_valid2)
plot_series(time_valid, rnn_forecast)
plt.show()
```
